## Lists

Modify mutable objects with in-place operators...

````
x = [1, 2, 3]
y = x
x.append(4)  # this modifies both lists 
print("x: ", x)
print("y: ", y)
````

... and avoid the issue by copying the object first.

````
x = [1, 2, 3]
y = x[:]  #  copy the whole list in memory
x.append(4)
print("x: ", x)
print("y: ", y)
````

Other operations with lists.

````
l = list(range(0,5))
print(l)

l = list(range(-10,-20,-1))
print(l)
````

## Control of flow

Examples of ```for``` loops.

````
a = ['a', 'b', 'c']

for elem in a:
    print(elem)

print(" ")

for i, elem in enumerate(a):
    print(i, elem)

print(" ")

for i in range(len(a)):
    print(i, a[i])
````

````
x = x.ravel()
th = 0.8
#print(x)

#Â how do I find the elements of x that are below a given threshold th?
z = x <= th
prob_z = np.mean(z)
print(prob_z)

th_values = np.linspace(0, 1, 100, endpoint=True)
print(th_values.shape)

prob_z = np.zeros(shape=th_values.shape)
print(prob_z)

for i, th in enumerate(th_values):
    prob_z[i] = (x <= th).mean()

plt.plot(th_values, prob_z)

````
