gaussian density cdf


```
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x_size1 = 10000

x = np.random.rand(x_size1, 1) * 20 - 10
# plt.plot(x, '.')
# plt.figure()

pdf_x = norm.pdf(x)
plt.plot(x, pdf_x, '.')
plt.figure()

cdf_x = norm.cdf(x)
plt.plot(x, cdf_x, '.')
```

```
th_values = np.linspace(0,1,x_size1, endpoint=True)
prob_z = np.zeros(th_values.shape)

for i, th in enumerate(th_values):
  prob_z[i] = (x <= th).mean()

plt.plot(th_values, prob_z, '.')
```
