```
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.random.randn(100, 1)
plt.plot(x, '.')
plt.figure()

pdf_x = norm.pdf(x)
plt.plot(x, pdf_x, '.')
         

```

```
th_values = np.linspace(0,1,100, endpoint=True)
prob_z = np.zeros(x.shape)

for i, th in enumerate(th_values):
  z = x <= th
  prob_z[i] = np.mean(z)

plt.plot(th_values, prob_z, '.')
```
